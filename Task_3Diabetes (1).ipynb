```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
dia=pd.read_excel("/content/diabetes_data.csv.xlsx")
```

```python
dia.head()
```

```python
dia.isnull().sum()
```

```python
LR=LogisticRegression()
```

```python
ind=dia[['weight','blood_glucose','bmi']]
dep=dia['risk_score']
```

```python
LR.fit(ind,dep)
```

```python
weight=int(input("enter the weight :"))
blood_glucose=int(input("enter the blood_glucose"))
bmi=int(input("enter the bmi"))
pred=LR.predict([[weight,blood_glucose,bmi]])
```

```python
print(pred)
```

```python
LR.score(ind,dep)
```

```python
from sklearn.metrics import accuracy_score
pval=LR.predict(ind)
accuracy_score(dep,pval)
```
